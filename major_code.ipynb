TRAIN CROP RECOMMENDATION SYSTEM

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("crop_remmendation_dataset.csv")

# Encode categorical data
le = LabelEncoder()
categorical_cols = ['Soil_Type', 'Region', 'Season',
                    'Irrigation_Type', 'Fertilizer_Used', 'Previous_Crop']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Features
X = df[['N','P','K',
        'Soil_pH','Soil_Moisture',
        'Temperature','Humidity','Rainfall',
        'Sunlight_Hours','Wind_Speed','Altitude',
        'Soil_Type','Region','Season',
        'Irrigation_Type','Fertilizer_Used','Previous_Crop']]

# Target
y = df['Recommended_Crop']

# Train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained successfully ✅")
```

YEILD PREDICTION


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Load data
yield_df = pd.read_csv("Crop Yeild Data.csv")

# Clean column names
yield_df.columns = yield_df.columns.str.strip()

# Encode categorical columns
le = LabelEncoder()
for col in ['Crop', 'Season', 'State']:
    yield_df[col] = le.fit_transform(yield_df[col])

# Features
X = yield_df[['Crop',
              'Season',
              'State',
              'Annual_Rainfall',
              'Avg_Temperature',
              'Area',
              'Fertilizer',
              'Pesticide']]

# Target
y = yield_df['Yield']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

print("Yield model trained successfully ✅")
```

PRICE PREICTION

```python
import pandas as pd

# Load dataset
price_df = pd.read_csv("commodity_price.csv")

# -------------------------------
# Step 1: Clean column names
# -------------------------------
price_df.columns = price_df.columns.str.strip()

# -------------------------------
# Step 2: Normalize crop names
# -------------------------------
price_df['Commodity'] = price_df['Commodity'].str.lower()

# -------------------------------
# Step 3: Convert date (optional but useful)
# -------------------------------
price_df['Arrival_Date'] = pd.to_datetime(price_df['Arrival_Date'], dayfirst=True)

# -------------------------------
# Step 4: Use latest prices (recommended)
# -------------------------------
latest_prices = price_df.sort_values('Arrival_Date').groupby('Commodity').tail(1)

# -------------------------------
# Step 5: Create price map
# -------------------------------
price_map = latest_prices.set_index('Commodity')['Modal_x0020_Price'].to_dict()

# -------------------------------
# Step 6: Price prediction function
# -------------------------------
def get_price(predicted_crop):
    crop = predicted_crop.lower()

    # Try exact match
    if crop in price_map:
        return price_map[crop]

    # Try partial match (IMPORTANT)
    for commodity in price_map:
        if crop in commodity:
            return price_map[commodity]
    return 1000
predicted_crop = "rice"

price = get_price(predicted_crop)

print("Predicted Crop:", predicted_crop)
print("Estimated Price:", price)
```

```python
def get_price(predicted_crop, state=None):
    crop = predicted_crop.lower()
    
    df = price_df.copy()
    
    if state:
        df = df[df['State'].str.lower() == state.lower()]
    
    df = df.sort_values('Arrival_Date').groupby('Commodity').tail(1)
    temp_map = df.set_index('Commodity')['Modal_x0020_Price'].to_dict()

    for commodity in temp_map:
        if crop in commodity:
            return temp_map[commodity]

    return 1000
```
